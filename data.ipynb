```python
from math import *
def getLocation():
    data=[]
    for dt in range(0,361):
        for dr in range(1, 129 * 3):        
            data.append([900 + int(dr * cos(dt / 180 * 2 * pi)), 870 + int(dr * sin(dt / 180 * 2 * pi))]) #中天社区
        for dr in range(1, 29 * 3):        
            data.append([1189 + int(dr * cos(dt / 180 * 2 * pi)), 591 + int(dr * sin(dt / 180 * 2 * pi))]) #云顶苑
        for dr in range(1, 1 * 3):        
            data.append([1264 + int(dr * cos(dt / 180 * 2 * pi)), 505 + int(dr * sin(dt / 180 * 2 * pi))]) #云顶苑4栋高层
        for dr in range(1, 52 * 3):        
            data.append([911 + int(dr * cos(dt / 180 * 2 * pi)), 569 + int(dr * sin(dt / 180 * 2 * pi))]) #俊仕苑
        for dr in range(1, 86 * 3):        
            data.append([407 + int(dr * cos(dt / 180 * 2 * pi)), 462 + int(dr * sin(dt / 180 * 2 * pi))]) #天骄苑
        for dr in range(1, 13 * 3):        
            data.append([150 + int(dr * cos(dt / 180 * 2 * pi)), 274 + int(dr * sin(dt / 180 * 2 * pi))]) #天骄苑38栋高层
        for dr in range(1, 39 * 3):        
            data.append([622 + int(dr * cos(dt / 180 * 2 * pi)), 607 + int(dr * sin(dt / 180 * 2 * pi))]) #爱唯大地幼儿园
        for dr in range(1, 58 * 3):        
            data.append([611 + int(dr * cos(dt / 180 * 2 * pi)), 1010 + int(dr * sin(dt / 180 * 2 * pi))]) #金爵苑
        for dr in range(1, 55 * 3):        
            data.append([1136 + int(dr * cos(dt / 180 * 2 * pi)), 935 + int(dr * sin(dt / 180 * 2 * pi))]) #领秀苑
    return data
```

```python
from pyheatmap.heatmap import HeatMap
data = getLocation()
heat = HeatMap(data)
heat.clickmap(base = 'image/中天社区.jpg', save_as = 'image/click.png')
heat.heatmap(base = 'image/中天社区.jpg', save_as = 'image/heat.png')
```

```python
import tkinter as tk

window = tk.Tk()
window.resizable(0,0)

image = tk.PhotoImage(file='image/heat.png')
image = image.subsample(2, 2)
canvas = tk.Canvas(window, width=image.width(), height=image.height())
canvas.pack()

canvas.create_image(0, 0, anchor=tk.NW, image=image)
popup = tk.Toplevel()
popup.overrideredirect(True)
name = tk.StringVar()
name.set('')
label = tk.Label(popup, textvariable = name).pack()

def on_mouse_move(event):
    # 获取鼠标坐标
    x, y = event.x, event.y
    # 判断鼠标是否在范围内
    if 431 < x < 471 and 415 < y < 455:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('中天社区: 66\n监督组: 5\n综合组: 44\n金山街道综合网格指挥中心: 14')
        canvas.create_oval(451 - 159, 435 - 159, 451 + 159, 435 + 159, tag = 'circle_highlight', outline = 'red')
    elif 574 < x < 614 and 275 < y < 315:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('云顶苑: 29')
    elif 611 < x < 651 and 232 < y < 272:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('云顶苑4栋高层: 1')
    elif 436 < x < 476 and 264 < y < 304:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('俊仕院: 52')
    elif 183 < x < 223 and 211 < y < 251:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('天骄苑: 86')
    elif 56 < x < 96 and 117 < y < 157:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('天骄苑38栋高层: 13')
    elif 291 < x < 331 and 283 < y < 323:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('爱唯大地幼儿园: 39')
    elif 286 < x < 326 and 486 < y < 526:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('金爵苑: 58')
    elif 549 < x < 589 and 448 < y < 488:
        popup.deiconify()
        popup.geometry('+{}+{}'.format(x + 15 + window.winfo_x(), y + 5 + window.winfo_y()))
        name.set('领秀苑: 55')
    else:
        popup.withdraw()
        try:
            canvas.delete('circle_highlight')
        except (UnboundLocalError):
            pass
        
canvas.bind('<Motion>', on_mouse_move)

window.mainloop()
```
